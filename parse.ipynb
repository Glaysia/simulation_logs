```python
import glob
import json

# Load all JSON files from current directory into a list of dicts
dicts_list = [json.load(open(fp, 'r', encoding='utf-8')) for fp in glob.glob('*.json')]
```

```python
len(dicts_list)
```

```python
new = []
fail =[]
for i in dicts_list:
    dd = i['progress_dict']
    go = dd[list(dd).pop()][0] =='coreloss_project'
    # print(dd[list(dd).pop()][0])
    if go:
        new.append(i)
    else:
        fail.append(i)
len(new)
```
